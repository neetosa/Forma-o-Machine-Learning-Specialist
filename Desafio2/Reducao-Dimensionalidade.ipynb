# Importance of Dimensionality Reduction

**Dimensionality reduction is the process of reducing the number of random variables under consideration, by obtaining a set of principal variables.**

----------------------------------------
## Problem with high dimensional data?
----------------------------------------

  1. It can mean high computational cost to perform learning.
  2. It often leads to over-fitting when learning a model, which means that the model will perform well on the training data but poorly on test data.
  3. Data are rarely randomly distributed in high-dimensions and are highly correlated, often with spurious correlations.
  4. The distances between a nearest and farthest data point can become equidistant in high dimensions, that can hamper the accuracy of some distance-based analysis tools.

----------------------------------------
## Why do we need Dimensionality Reduction?
----------------------------------------

  1. Dimensionality reduction helps with these problems, while trying to preserve most of the relevant information in the data needed to learn accurate, predictive models.
  2. There are often too many factors on the basis of which the final prediction is done. These factors are basically variables called features.
  3. The higher the number of features, the harder it gets to visualize the training set and then work on it.
  4. Sometimes, most of these features are correlated, and hence redundant. This is where dimensionality reduction algorithms come into play.

----------------------------------------
## Importance of Dimensionality reduction?
----------------------------------------

  1. It reduces the time and storage space required.
  2. It helps Remove multi-collinearity which improves the interpretation of the parameters of the machine learning model.
  3. It becomes easier to visualize the data when reduced to very low dimensions such as 2D or 3D.
  4. It avoids the curse of dimensionality.
  5. It removes irrelevant features from the data, Because having irrelevant features in the data can decrease the accuracy of the models and make your model learn based on irrelevant features.

# Gray Scale

The computer processes images through three distinct channels: red, green, and blue. Each of these channels (red, green, and blue) is represented by an array.

three_d_array.png

For grayscale images, the result is a two-dimensional array with the number of rows and columns equal to the number of pixel rows and columns in the image. Low numeric values indicate darker shades and higher values lighter shades.

To obtain a grayscale image, we can apply a linear transformation to the RGB channels, given by the following expression.

$$\text{Gray Scale} = 0.299R + 0.587G + 0.114B$$

```python
# Open image

image_path = '/content/drive/MyDrive/EAFC25.jpg'

image = Image.open(image_path)

display(image)

print('Image format:', image.format)
print("Image mode:", image.mode)
print("Image size:", image.size)
```

```python
# Gray image

image_array = np.array(image)

height, width, _ = image_array.shape

gray_array = np.zeros((height, width), dtype = np.uint8)

for i in range(height):
  for j in range(width):
    R, G, B = image_array[i,j]

    gray_value = int(0.299*R+0.587*G+0.114*B)

    gray_array[i,j] = gray_value

gray_image = Image.fromarray(gray_array, mode = 'L')

display(gray_image)
```

```python
# Make an black and white image

threshold = 110

bw_array = np.zeros_like(gray_array, dtype = np.uint8)

bw_array[gray_array < threshold] = 0
bw_array[gray_array >= threshold] = 255

bw_image = Image.fromarray(bw_array, mode = 'L')

display(bw_image)
```
